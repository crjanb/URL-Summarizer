```python
import os
import streamlit as st
import pickle
import time
import langchain
from langchain_community.llms.ollama import Ollama
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import UnstructuredURLLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
```

```python
llm = Ollama(model="llama3", temperature=0.9)
```

```python
# (1) Load data
loaders = UnstructuredURLLoader(urls=[
    "https://kathmandupost.com/national/2025/03/10/rastriya-prajatantra-party-pays-rs100-000-fine-to-kathmandu-metropolitan-city",
    "https://en.setopati.com/political/164347"
])
data = loaders.load()
print(f"Number of documents loaded: {len(data)}")
```

```python
# (2) Split data to create chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
docs = text_splitter.split_documents(data)
print(f"Number of document chunks: {len(docs)}")
```

```python
# (3) Create embeddings using HuggingFaceEmbeddings and build a FAISS index
# Using the all-MiniLM-L6-v2 model, which is a free, open-source option.
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorindex = FAISS.from_documents(docs, embeddings)

# Save the FAISS vector index locally for reuse
file_path = "vector_index.pkl"
with open(file_path, "wb") as f:
    pickle.dump(vectorindex, f)

# Optionally, load the vector index if it already exists
if os.path.exists(file_path):
    with open(file_path, "rb") as f:
        vectorindex = pickle.load(f)
```

```python
# (4) Retrieve similar documents for a given query and generate an answer using Llama 3
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorindex.as_retriever())

query = "Why RPP paid fine?"
langchain.debug = True

result = chain({"question": query}, return_only_outputs=True)
print(result)
```
